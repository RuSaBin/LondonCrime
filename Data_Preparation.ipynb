```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
crime_rate = pd.read_csv('london_crime_by_lsoa.csv', encoding = 'utf-8', sep = ",")
```

```python
crime_rate = crime_rate[crime_rate.year != (2008 or 2009 or 2010)]  #delete data from 2008 to 2010
```

```python


crime_rate = crime_rate[crime_rate.value != 0] #delete rows where 'value' is 0
```

```python
crime_rate.shape[0]
```

```python
crime_rate.major_category.value_counts().keys() #columns in new data frame
```

```python
len(crime_rate.borough.value_counts().keys())
```

```python
index = range(len(crime_rate.borough.value_counts().keys())) #create empty data frame

columns = ['borough','Theft and Handling', 'Violence Against the Person', 'Criminal Damage', 'Burglary', 'Drugs', 'Robbery', 'Other Notifiable Offences', 'Total', 'Proportion']


df = pd.DataFrame(index=index, columns=columns)
df = df.fillna(0) 
```

```python
df.borough=crime_rate.borough.value_counts().keys().sort_values().values #fill borough names
    
```

```python
df.loc[df.borough=='Barking and Dagenham', 'Theft and Handling'] = crime_rate[(crime_rate.borough == 'Barking and Dagenham')&(crime_rate.major_category == 'Theft and Handling')]['value'].sum(
)
```

```python
crime_rate[(crime_rate.borough == 'Barking and Dagenham')&(crime_rate.major_category == 'Theft and Handling')]['value'].sum(
)
```

```python
# sums for each crime category and total 

for i in range(df.shape[0]):
    name = df.loc[i].borough
    for j in range(len(df.columns)-2):
        cat = df.columns[j+1]
        df.loc[df.borough==name, cat] = crime_rate[(crime_rate.borough == name)&(crime_rate.major_category == cat)]['value'].sum()
        if cat == 'Total':
            df.loc[df.borough==name, cat] = crime_rate[crime_rate.borough == name]['value'].sum()
            

```

```python
# proportions of totals
sum_of_totals = df.Total.sum()
for i in range(df.shape[0]):
    name = df.loc[i].borough
    total = df.loc[df.borough==name, 'Total']
    df.loc[df.borough==name, 'Proportion'] = round(total/sum_of_totals*100, ndigits = 2)
```
